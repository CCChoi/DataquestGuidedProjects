```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")
star_wars.head()
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()
```

```python
yes_no = {"Yes": True, "No": False}


star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
star_wars.head()
```

```python
episode_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}


for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(episode_map)

star_wars = star_wars.rename(columns={    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
     })                                 
        

    
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={  
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "rank_1",
    "Unnamed: 10": "rank_2",
    "Unnamed: 11": "rank_3",
    "Unnamed: 12": "rank_4",
    "Unnamed: 13": "rank_5",
    "Unnamed: 14": "rank_6"
     })      

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

```python
means = star_wars[star_wars.columns[9:15]].mean()

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

As predicted in the beggining of the instructions, Episode V is the the most popular.  Additionally, all of the original trilogy (IV, V, VI) scored much better than the newer three.


```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

In addition being more popular on average, the original three movies had been seen more often.  Though the difference is not as significant.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()
plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

More males than females saw the newer three movies but on average liked the movies less.  Phantom Menace was much better received by female viewers, but still was liked less than the original trilogy.
